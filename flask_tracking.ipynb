```python
locs = ['ALT', 'AMY', 'BRW','CGO',
        'HFM','KUM','LEF','MHD','MLO',
        'NWR','SMO','SPO','THD','C14','UCI']
if 'stations' in globals():
    del stations
    print('station variable deleted')
stations = ft.dalist(locs)
stations.view('HFM')
time_tag()
```

```python
def external_save(da_list, location='./'):
        q = input('Save? (y/n)')
        if q == 'y':
            for da in da_list.da_list:
                filename = da.name+'.'+time_tag()+'.nc'
                path = location+da.name+'/'+filename
                recent_path = location+da.name+'/'+da.name+'.nc'
                da = da.astype(str)
                da.to_netcdf(path)#, mode='w') # specific dated instance
                print(path, 'saved')
                da.to_netcdf(recent_path)#, mode='w') # generic "current" file overwrite
                print(recent_path, 'saved')
```

```python
external_save(stations, location='/Users/matthewgentry/Google_Drive/flasktracking/')
```
